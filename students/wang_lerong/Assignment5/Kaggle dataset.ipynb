```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
alcohol = pd.read_csv("student-mat.csv")
```

```python
alcohol
```

```python
alcohol.hist(column="Walc", color="blue") 
plt.suptitle("Histogram of weekend alcohol consumption for student in math course", 
             x=0.5, y=1.05, ha='center', fontsize='large')
plt.xticks(np.arange(np.min(alcohol["Walc"]), np.max(alcohol["Walc"]) + 1))
plt.xlabel("weekend alcohol consumption")
plt.ylabel("frequency")
```

```python
plt.scatter(alcohol['absences'],
            alcohol['Dalc']) 
plt.title('Scatter plot of the relationship between weekday alcohol consumption and school absences')
plt.ylabel('weekday alcohol consumption')
plt.xlabel('absences')
```
