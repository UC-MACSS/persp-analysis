## Exploration Write-Up

The General Social Survey (GSS) collects information on the demographics and attributes of the American Society. By conducting exploration data analysis of this dataset, we can identify hidden patterns of some subgroups of the population, which might be further studied and add a contribution to the literature. 

One research question that has received a great deal of attention is that whether peoples' happiness is associated with their ideology. Previous studies suggest that conservatives are more likely to report a higher level of happiness in the United States. However, using a different method of measurement of happiness, Wojcik, Sean P., et al.(2015) found out that this higher level of happiness is fully mediated by conservatives' self-enhancing style. As is described by Wojcik, Sean P., et al.(2015), conservatism has been characterized as an ideology grounded in ego defensiveness, enhanced sensitivity to negativity, and personality dimensions related to defensive forms of motivated social cognition. Hence, it is reasonable to assume that self-enhancement is more prominent among conservatives than liberals. They further state that self-enhancement is more pronounced among religious people. Since the data contains individuals' religions, we can examine the effect of self-enhancement on self-reported happiness: if self-enhancement does have a positive effect on self-reported happiness, the level of reported happiness should be lower for the individuals without religion, who theoretically have a lower degree of self-enhancement. Then, we can compare the individuals' self-reported happiness for those with/without religion among groups with different political views. 

Since our analysis is purely based on data exploration, the conclusion might not be rigorous: though the question of interest focuses on identifying the causal relationship between self-enhancement and self-reported happiness, here, we investigate their relationship rather than causality. Statistical models have to be carefully constructed in order to derive a causal inference. However, our analysis can provide strong incentives for future research on related topics. 

First, we plot the histogram of the self-reported happiness over the entire population. Instead of a psychological scale, the survey participants choose their level of happiness from three broad categories: very happy, pretty happy and not too happy. From the histogram, we can see that the majority of the survey participants (55.7%) consider themselves being "pretty happy". 33% of them chose "very happy", while only 14.1% reported themselves being "not too happy". As for political views, 38% of the participants classified themselves as "moderate", while the proportion of conservative-inclined individuals are slightly higher than the liberal-inclined individuals.

Then, I divided the data into two groups: individuals with/without religion. The size of the two groups are different: 1587 observations are presented in the "with religion" group while the "without religion" group has only 387 observations. According to above reasoning, the general level of happiness should be higher for religious people than non-religious people. From the pie chart, we can see that, though the proportion of individuals reporting "very happy" is less for religious people (13.4% vs. 17.1% for non-religious), the proportion of individuals feels "happy" is indeed greater for the religious group (86.4% vs 82.9% for non-religious). This is consistent with previous hypothesis.

```python
plt.figure(figsize=(12,10))
plt.subplot(2,1,1)
sns.countplot(x="happy", hue="polviews", data=non_relig, hue_order = 
              ["ExtrmCons", "Conserv", "SlghtCons", "Moderate","SlghtLib","Liberal","ExtrmLib"])
plt.ylabel(r'Frequency')
plt.title(r'No-religion: Histogram of Self-reported Happiness by Political Views')
plt.subplot(2,1,2)
sns.countplot(x="happy", hue="polviews", data=relig, order = ["NOT TOO HAPPY", "PRETTY HAPPY", "VERY HAPPY"],
hue_order = ["ExtrmCons", "Conserv", "SlghtCons", "Moderate","SlghtLib","Liberal","ExtrmLib"])
plt.ylabel(r'Frequency')
plt.title(r'Religion: Histogram of Self-reported Happiness by Political Views')
```

The graph demonstrates the relationship between religion and self-reported happiness. Among people reporting "very happy", the relative frequencies of all conservative-inclined categories for non-religious people are significantly lower than religious people. This pattern is also salient among people responded "pretty happy". Comparing to non-religious people, the proportion of liberals are dramatically lower in the religious group, while the proportion of conservatives are much higher. Because there are few people reporting "not too happy", the pattern is not easily observable in this group. All these foundings support that our hypothesis: political conservatives are happier partially due to their inherited propensity of self-enhancement.

To emphasize, since it is difficult to prove the correctness of hypotheses in social science subjects, our analysis does not intend to identify the causal mechanism between political ideology and happiness with constructing a statistical model. Instead, it serves for future study on related topics. Based on above analysis, the hypothesis that self-enhancement acts as a channel to affect people's self-reported happiness is consistent with the patterns observed from the data. Future study can be conducted by controlling confounding factors. Due to the limitation of our analysis, it is also important for researchers to ask themselves the question: is there a channel rather than self-enhancement through which religion/political ideology can change people's feeling of happiness?
