### Statistical estimates

```python
describe
```

Consider sample mean. With age which has high count, std is quite high because it is pretty uniform distribution. But due to high sample size (count), the 95% confidence interval is narrow:

```python
import math
```

```python
(describe['age'].loc['mean'] - 1.96 * describe['age'].loc['std']/math.sqrt(describe['age'].loc['count']), describe['age'].loc['mean'] + 1.96 * describe['age'].loc['std']/math.sqrt(describe['age'].loc['count']))
```

Our confidence interval of average age is very wide
